### **Exploratory Data Analysis (EDA)**

**Exploratory Data Analysis (EDA)** is a crucial step in the data analysis process where analysts use statistical graphics and other data visualization methods to explore the dataset's structure and characteristics.

Common techniques used in EDA include:

- **Descriptive Statistics**: Summarizing the central tendency, dispersion, and shape of a dataset’s distribution (e.g., mean, median, mode, standard deviation).
- **Visualizations**: Creating plots such as histograms, box plots, scatter plots, and bar charts to visually inspect data.
- **Data Cleaning and Transformation**: Addressing missing values, handling outliers, and transforming variables if needed.

### **Univariate Analysis**

**Univariate Analysis** refers to the examination of a single variable. The objective is to describe and summarize the variable’s distribution and key characteristics. It is a fundamental part of EDA and helps in understanding the properties of individual variables before moving on to more complex multivariate analyses. Key aspects of univariate analysis include:

1. **Summary Statistics**: Calculating measures such as:
   - **Central Tendency**: Mean, median, mode.
   - **Dispersion**: Range, variance, standard deviation, interquartile range (Q3-Q1).
   - **Shape**: Skewness, kurtosis.

2. **Frequency Distribution**: Counting the occurrences of each unique value or range of values, often presented in tables or histograms.

3. **Visualization**: Using plots to visualize the distribution of the variable:
   - **Histogram**: Displays the frequency distribution of a continuous variable.
   - **Box Plot**: Shows the distribution, central value, and variability of a variable, highlighting outliers.
   - **Bar Chart**: Used for categorical variables to show the count of each category.

### Example of EDA and Univariate Analysis

Consider a dataset containing the ages of individuals in a survey. In a univariate analysis of the "age" variable:

1. **Summary Statistics**:
   - Mean age: 35 years.
   - Median age: 34 years.
   - Standard deviation: 10 years.

2. **Frequency Distribution**:
   - Ages 20-30: 40 individuals.
   - Ages 31-40: 60 individuals.
   - Ages 41-50: 30 individuals.

3. **Visualization**:
   - **Histogram**: Shows a peak around the 30-40 age range.
   - **Box Plot**: Indicates the central 50% of ages are between 28 and 42, with a few outliers above 60.

By performing EDA and univariate analysis, analysts gain initial insights that guide further in-depth analysis and model building.

# **Matplotlib and Seaborn Libraries in Python**

**Matplotlib** and **Seaborn** are two of the most commonly used libraries for data visualization in Python. Each has its own strengths and is often used together to create informative and aesthetically pleasing plots.

## **Matplotlib**

**Matplotlib** is a comprehensive library for creating static, animated, and interactive visualizations in Python. It is highly customizable and is widely used for producing publication-quality figures.

**Key Features:**
- **Versatility**: Matplotlib can create a wide variety of plots including line charts, scatter plots, bar charts, histograms, pie charts, and 3D plots.
- **Customization**: Every aspect of a plot can be customized, from colors and line styles to fonts and labels.
- **Integration**: Matplotlib integrates well with other libraries such as NumPy and Pandas, making it easier to plot data directly from these structures.
- **Interactive Plots**: Using `matplotlib.pyplot`, users can create interactive plots in Jupyter notebooks.

**Example:**
```python
import matplotlib.pyplot as plt

# Simple line plot
x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

## **Seaborn**

**Seaborn** is built on top of Matplotlib and provides a high-level interface for drawing attractive and informative statistical graphics. It simplifies complex visualizations by providing default themes and color palettes.

**Key Features:**
- **Statistical Plots**: Seaborn includes built-in themes for statistical plots such as box plots, violin plots, pair plots, and regression plots.
- **Ease of Use**: Seaborn is designed to work well with Pandas data structures, making it easy to create visualizations directly from DataFrames.
- **Aesthetic Defaults**: Seaborn's default styles and color palettes make it easy to create visually appealing plots with minimal customization.
- **Data Visualization**: It is particularly powerful for visualizing distributions of data and relationships between variables.

**Example:**
```python

import pandas as pd

# Example DataFrame

    'Age': [32, 28, 45, 36, 52],
    'Salary': [90000, 65000, 150000, 60000, 200000]

sns.barplot(x='Age', y='Salary', data=df)
plt.title('Bar Plot')

```

### Differences and When to Use Each

- **Customization**: Matplotlib offers extensive customization options, making it suitable for creating detailed and complex plots where specific control over plot elements is needed.
- **Ease of Use**: Seaborn abstracts many of Matplotlib's complexities and provides simpler syntax for creating common statistical plots. It is ideal for quick and attractive visualizations.
- **Statistical Visualizations**: Seaborn excels in creating informative and attractive statistical graphics with minimal effort, making it a go-to choice for data exploration and analysis.
- **Integration**: Both libraries work well with other data manipulation libraries like Pandas, but Seaborn has tighter integration and more intuitive handling of DataFrames.

# **Histograms**

A histogram is a graphical representation of the distribution of numerical data. It consists of a series of adjacent rectangles (bins), each representing a range of data values, with the height of each rectangle corresponding to the frequency or count of data points within that range.

## **Key Features of Histograms:**

**Bins:** The data range is divided into intervals called bins. Each bin represents a range of data values. The number of bins and their width can be adjusted to provide different levels of granularity in the visualization.


**Frequency/Count:** The height of each rectangle (bin) represents the frequency or count of data points falling within that bin's range. It shows how many data points fall into each bin.

**Density:** In some cases, histograms can display the density of data points rather than their counts. This can be useful when comparing distributions with different numbers of total data points.

Below code is creating a histogram for the 'Age' column of salary  DataFrame using Seaborn's histplot() function.

````python
# Summary statistics for 'Age'
summary_stats = salary_df['Age'].describe()
print(summary_stats)

# Histogram for 'Age'
plt.figure(figsize=(8, 6))
sns.histplot(salary_df['Age'], bins=10, kde=True)
plt.title('Histogram of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
````


Here's a breakdown of what each part of the code does:

1. **Computing Summary Statistics**:
   - `summary_stats = salary_df['Age'].describe()`: Computes summary statistics (mean, standard deviation, minimum, maximum, etc.) for the 'Age' column using the `describe()` function in pandas. This function provides descriptive statistics for numerical columns in the DataFrame.

2. **Creating the Histogram**:
   - `plt.figure(figsize=(8, 6))`: Creates a new figure with a specified figure size of 8x6 inches.
   - `sns.histplot(salary_df['Age'], bins=10, kde=True)`: Creates a histogram of the 'Age' column using Seaborn's `histplot()` function.
     - `salary_df['Age']` specifies the data to be plotted, which is the 'Age' column from the DataFrame `salary_df`.
     - `bins=10` specifies the number of bins for the histogram. In this case, there will be 10 bins.
     - `kde=True` overlays a kernel density estimate (KDE) plot on top of the histogram, showing the estimated probability density function.
   - `plt.title('Histogram of Age')`: Adds a title to the plot with the text 'Histogram of Age'.
   - `plt.xlabel('Age')`: Adds a label to the x-axis with the text 'Age'.
   - `plt.ylabel('Frequency')`: Adds a label to the y-axis with the text 'Frequency'.
   - `plt.show()`: Displays the plot.


````python
# Summary statistics for 'Age'
summary_stats = salary_df['Age'].describe()
mean_age = summary_stats['mean']
std_age = summary_stats['std']
min_age = summary_stats['min']
max_age = summary_stats['max']

print(summary_stats)

# Histogram for 'Age'
plt.figure(figsize=(10, 6))
sns.histplot(salary_df['Age'], bins=10, kde=True)

# Adding mean, min, max, and std lines
plt.axvline(mean_age, color='r', linestyle='--', linewidth=2, label=f'Mean: {mean_age:.2f}')
plt.axvline(min_age, color='g', linestyle='-', linewidth=2, label=f'Min: {min_age}')
plt.axvline(max_age, color='b', linestyle='-', linewidth=2, label=f'Max: {max_age}')
plt.axvline(mean_age - std_age, color='purple', linestyle='--', linewidth=2, label=f'Mean - 1 Std: {mean_age - std_age:.2f}')
plt.axvline(mean_age + std_age, color='purple', linestyle='--', linewidth=2, label=f'Mean + 1 Std: {mean_age + std_age:.2f}')

# Annotating the lines
plt.text(mean_age, plt.gca().get_ylim()[1]*0.9, f'Mean: {mean_age:.2f}', color = 'r', ha='center')
plt.text(min_age, plt.gca().get_ylim()[1]*0.8, f'Min: {min_age}', color = 'g', ha='center')
plt.text(max_age, plt.gca().get_ylim()[1]*0.8, f'Max: {max_age}', color = 'b', ha='center')
plt.text(mean_age - std_age, plt.gca().get_ylim()[1]*0.7, f'Std: {mean_age - std_age:.2f}', color = 'purple', ha='center')
plt.text(mean_age + std_age, plt.gca().get_ylim()[1]*0.7, f'Std: {mean_age + std_age:.2f}', color = 'purple', ha='center')

plt.title('Histogram of Age with Summary Statistics')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.legend()
plt.show()
````
