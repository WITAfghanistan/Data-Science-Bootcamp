# **Data Cleaning**

*pd.to_numeric()* is a pandas function used to convert the specified column to numeric data type. The parameter errors='coerce' means that if there are any errors during conversion, they will be replaced with NaN (Not a Number).

*.str.replace(',', '')* is used to remove commas from the international_students column, as commas can interfere with converting the column to a numeric data type.

*.astype(float, errors='ignore')* converts the modified international_students column to a floating-point data type while ignoring errors. This is because some values might not be convertible to float, and those will remain as strings.

```python
uni['rank_display'] = pd.to_numeric(uni['rank_display'], errors='coerce')
uni['score'] = pd.to_numeric(uni['score'], errors='coerce')
uni['student_faculty_ratio'] = pd.to_numeric(uni['student_faculty_ratio'], errors='coerce')
uni['international_students'] = uni['international_students'].astype(str).str.replace(',', '').astype(float, errors='ignore')
uni['faculty_count'] = pd.to_numeric(uni['faculty_count'], errors='coerce')
```

This part of the code fills any missing values in the specified columns with appropriate methods, ensuring data consistency and integrity.

*.fillna()* is a pandas function used to fill missing values. It takes the median value of each column as the argument to fill the missing values. The median is a robust measure of central tendency that is less affected by outliers compared to the mean.

The *inplace=True* parameter ensures that the modifications are made directly to the DataFrame uni without the need for reassignment.

```python
uni['rank_display'].fillna(uni['rank_display'].median(), inplace=True)
uni['score'].fillna(uni['score'].median(), inplace=True)
uni['student_faculty_ratio'].fillna(uni['student_faculty_ratio'].median(), inplace=True)
uni['international_students'].fillna(uni['international_students'].median(), inplace=True)
uni['faculty_count'].fillna(uni['faculty_count'].median(), inplace=True)
```

# **Handling duplicates**

*drop_duplicates()* : This is a method provided by pandas to remove duplicate rows from the DataFrame. By default, it considers all columns to identify duplicates and keeps the first occurrence of each duplicate row.

```python
uni = uni.drop_duplicates(['country'])
uni
```

The code *uni = uni.drop_duplicates(['country'])* removes duplicate rows from the uni DataFrame based on the values in the country column. This means that only the first occurrence of each unique country will be kept, and any subsequent rows with the same country value will be dropped.

# **Average Rank and Score by Year**

```python
average_rank_by_year = uni.groupby('year')['rank_display'].mean()
average_score_by_year = uni.groupby('year')['score'].mean()

plt.figure(figsize=(12, 6))
sns.lineplot(x=average_rank_by_year.index, y=average_rank_by_year.values, marker='o', label='Average Rank')
sns.lineplot(x=average_score_by_year.index, y=average_score_by_year.values, marker='o', label='Average Score', color='red')
plt.title('Average Rank and Score by Year')
plt.xlabel('Year')
plt.ylabel('Average Value')
plt.legend()
plt.show()
```

"*average_rank_by_year*" and "*average_score_by_year*" are calculated by grouping the uni DataFrame by the 'year' column and calculating the mean of 'rank_display' and 'score' columns, respectively.

*plt.figure(figsize=(12, 6))* creates a figure object with a specified size of 12x6 inches for the line plot.
*sns.lineplot()* creates line plots for average rank and average score by year using seaborn. It plots the values against the year index, with markers ('o') for data points. The label parameter is used to specify the legend labels for each line plot.

*plt.title('Average Rank and Score by Year'), plt.xlabel('Year')*, and *plt.ylabel('Average Value')* set the title, x-axis label, and y-axis label of the plot, respectively.

*plt.legend()* displays the legend on the plot to differentiate between average rank and average score lines.
*plt.show()* displays the plot.

**Chart: Average Rank and Score by Year**

Explanation:

Lines Representing Average Values: The chart shows two lines:
Average Rank (blue): The average rank of universities by year.
Average Score (red): The average score of universities by year.

The average rank has been increasing over the years, indicating that universities are generally getting lower rankings (higher numerical values).
The average score, however, remains relatively stable with slight fluctuations.

An increasing average rank suggests that the overall competitiveness or the number of universities being ranked might have increased over the years.
The stable average score indicates that while the ranks may vary, the quality (as represented by the score) has remained fairly consistent.
